3) Сравните построенные приближения между собой (по точности) и с "классическим" приближением $32\Bigl(({\rm year} - 1820)/100\Bigr)^2$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Созданние массивов с данными
years = [-1000,  -900,  -800,  -700,  -600,  -500,  -400,  -300,
         -200,  -100,     0,   100,   200,   300,   400,   500,
         600,   700,800,   900,  1000,  1100,  1200,  1300,
         1400,  1500,  1600, 1700,  1710,  1720,  1730,  1740,
         1750,  1760,  1770,  1780, 1790,  1800,  1810,  1820,
         1830,  1840,  1850,  1860,  1870, 1880,  1890,  1900,
         1910,  1920,  1930,  1940,  1950,  1960, 1970,  1980,  1990,  2000]

deltaT = [25400, 23700, 22000, 20400, 18800, 17190, 15530, 14080, 12790,
          11640, 10580, 9600, 8640,7680,6700,5710,4740,3810,2960,2200,
          1570,1090,740, 490, 320, 200, 120, 9, 10, 11, 11, 12, 13, 15,
          16, 17, 17, 14, 13, 12, 8, 6, 7, 8, 2, -5, -6, -3, 10, 21, 24,
          24, 29, 33, 40, 51, 57, 65]
```

Создание первого графика и ввод данных:

```python
plt.figure(figsize=(14.55, 4.85))
# Создание первого графика
plt.subplot(1, 2, 1)
plt.xlabel("year")
plt.ylabel("deltaT")
plt.title("Все данные")
# Ввод данных в первый график
plt.plot(years, deltaT, 'g')
pass
```

Создание второго графика, который содержит данные после 1700 года и ввод данных (от 28 индекса):

```python
# Создание второго графика, который содержит данные после 1700 года
plt.subplot(1, 2, 2)
plt.xlabel("year")
plt.ylabel("deltaT")
plt.title("Данные после 1700 года")
# Ввод данных во второй график, от 28 индекса, так как 1700 год находится именно там
plt.plot(years[28:60], deltaT[28:60], 'y')
pass
```

Надо выбрать, с какой даты выполнять апроксимацию (в моем случае с 1800 года):

```python
x = years[37:58]
y = deltaT[37:58]
```

Создание графика апроксимации, расстановка точек и вычисление многочленов.
<br><br>
Вычисление "Классического" приближения и добавление его на график:

```python
plt.figure(figsize=(16.55, 6.85))
plt.xlabel("Year")
plt.ylabel("deltaT")
plt.title("Аппроксимация данных")

plt.scatter(x, y, c='green', s=20, alpha=0.5,
            cmap='viridis')

for r in range(1, 5):
    func = np.polyfit(x, y, r)
    x1 = np.linspace(x[0], x[20], 50)
    y1 = np.polyval(func, x1)
    deltaFunc = np.poly1d(func)
    print("Многочлен %i степени: " % deltaFunc.order)
    print(deltaFunc)
    plt.plot(x1, y1, '', label="Приближение многочлена %i степени" % deltaFunc.order)
n = len(x)
array = []

for i in range(0, n):
    d = int(32*((x[i]-1820)/100)**2)
    array.append(d)

plt.plot(x, array, '', label="'Классическое' приближение" )
plt.legend()
pass
```

<h1>Вывод:</h1>
<br>
Приближение многочленов степени 1, 2, 3 оказалось более точным, чем приближение многочлена 4 степени. Также <b>"Классическое"</b> приближение оказалось <b>менее</b> точным, чем приближение многочленов степени 2, 3, 4, но <b>более</b> точным, чем приближение 1 степени.
